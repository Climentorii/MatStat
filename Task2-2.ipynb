```python
import numpy as np

ma = np.random.rand(25)
ma = -np.log(1 - ma)
ma.sort()

print(ma)
```

```python
import numpy as np
import statistics as st

print("Мода = ", st.multimode(ma))
print("Медиана = ", st.median(ma))
print("Размах = ", ma[24] - ma[0])

x_mid = aplha_1 = (np.sum(ma)) / 25
alpha_2 = sum([i * i for i in ma]) / 25
mu_3 = sum([(i - x_mid) ** 3 for i in ma]) / 25
mu_2 = alpha_2 - aplha_1 ** 2
asymmetry_k = mu_3 / (mu_2)**1.5

print("Альфа 1 = ", x_mid)
print("Альфа 2 = ", alpha_2)
print("Мю 3 = ", mu_3)
print("Мю 2 = ", mu_2)
print("Коэффициент асимметрии = ", asymmetry_k)
```

```python
import statistics as st
import matplotlib.pyplot as plt
import math
from scipy. stats import norm
from scipy.stats import bootstrap


var = np.arange(0, 5, 0.01)

def bootstrap (data, n_bootstrap_samples = 1000):
  return [np.random.choice(data, size=len(data), replace = True) for _ in range(n_bootstrap_samples)]

bootstrap_ma = bootstrap(ma)
bootstrap_rslt = []
for i in bootstrap_ma:
  bootstrap_rslt.append(i.mean())

distance = int(1 + math.log(1000, 2))

length_ = (max(bootstrap_rslt) - min(bootstrap_rslt)) / distance

plt.plot (var, norm.pdf(var, 1, 1 / 5))
plt.hist (bootstrap_rslt, bins = distance, weights=np.ones_like(bootstrap_rslt) / 1000 / length_)

plt.show()
```

```python
import statistics as st
import matplotlib.pyplot as plt
import math
from scipy. stats import norm
from scipy.stats import bootstrap


def bootstrap_for_asymmetry (data, n_bootstrap_samples = 1000):
  return [np.random.choice(data, size=len(data), replace = True) for _ in range(n_bootstrap_samples)]

bootstrap_asymmetry_ma = bootstrap_for_asymmetry(ma)
bootstrap_rslt_a = []
for j in bootstrap_asymmetry_ma:
  x_mid = aplha_1 = (sum(j)) / 25
  alpha_2 = sum([i * i for i in j]) / 25
  mu_3 = sum([(i - x_mid) ** 3 for i in j]) / 25
  mu_2 = alpha_2 - aplha_1 ** 2
  asymmetry_k = mu_3 / mu_2**1.5

  bootstrap_rslt_a.append(asymmetry_k)

distance = int(1 + math.log(1000, 2))

length_a = (max(bootstrap_rslt_a) - min(bootstrap_rslt_a)) / distance

plt.hist (bootstrap_rslt_a, bins = distance, weights=np.ones_like(bootstrap_rslt_a) / 1000 / length_a)

plt.show()
```
